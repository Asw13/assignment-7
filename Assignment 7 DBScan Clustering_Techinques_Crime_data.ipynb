```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
# Import Dataset
crime=pd.read_csv('crime_data.csv')
crime
```

```python
crime.info()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime
```

```python
# summary of the numeric features of the dataset
crime.describe().round()
```

```python
# distribution of the features - boxplot as a 2x3 grid of subplots
crime.plot(kind='box', subplots=True, layout=(2,3),figsize=(10,10))
plt.tight_layout();
```

```python
# scale the features using standard scaler from sklearn
scaler = StandardScaler()
crime_scaled = scaler.fit_transform(crime)
crime_scaled = pd.DataFrame(crime_scaled, columns=crime.columns)
crime_scaled
```

```python
crime_scaled.describe().round(2)
```

```python
# Instantiate a DBSCAN model object and fit it to the scaled data - take some values for eps and min_samples
# eps = 1, min_samples = 4
dbscan = DBSCAN(eps=1, min_samples=4)
```

```python
# get the labels of the clusters
labels = dbscan.fit_predict(crime_scaled)
labels
```

```python
# add the labels to the original dataset as a new feature
crime['cluster_labels'] = labels
crime.head()
```

```python
# get the cluster profiles - mean values of the features
crime.groupby('cluster_labels').mean().round(2)
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime['cluster_labels'],crime['UrbanPop'], c=dbscan.labels_) 
```
